## here we want our model to predect a next digit in series 

```python
import numpy as np

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
data = [[[i+j/400] for i in range(5)]for j in range(100)]
target = [(i+5)/400 for i in range(100)]
```

```python
data = np.array(data,dtype = float)
target = np.array(target,dtype = float)
```

```python
data.shape
```

```python
target.shape
```

```python
from sklearn.model_selection import train_test_split 

x_train,x_test,y_train,y_test = train_test_split(data,target,test_size = 0.2,random_state = 4)
```

```python
model = Sequential()
```

```python
model.add(LSTM((1),batch_input_shape = (None,5,1),return_sequences = False))
```

```python
model.compile(loss='mean_absolute_error',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train, epochs= 1000 ,validation_data=(x_test,y_test))
```

```python
results = model.predict(x_test)
```

```python
import matplotlib.pyplot as plt

plt.scatter(range(20),results,c='r')
plt.scatter(range(20),y_test,c='g')

plt.show()
```

```python
plt.plot(history.history['loss'])
plt.show()
```
